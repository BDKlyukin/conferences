# MSCI

## Volatility

```python
# loading up the data
daily_MSCI_index_df = pd.read_excel(r'data\prev_calc\daily\daily_MSCI_index.xlsx')
```

```python
quarterly_volatility_df = pd.DataFrame()

for country, country_df in daily_MSCI_index_df.groupby('Country Name'):
    
    for year in country_df['Report Date'].dt.year.unique():
        year_country_df = country_df[country_df['Report Date'].dt.year == year]
        
        for quarter in year_country_df['Report Date'].dt.quarter.unique():
            quarter_country_df = year_country_df[year_country_df['Report Date'].dt.quarter == quarter]
    
            annualized_quarterly_returns_volatility = ((quarter_country_df['Value'] / quarter_country_df['Value'].shift(1)).dropna()-1).std() * np.sqrt(252)
            
            quarter_country_df = quarter_country_df[quarter_country_df['Report Date'] == max(quarter_country_df['Report Date'])]
            quarter_country_df['Value'] = annualized_quarterly_returns_volatility
            quarterly_volatility_df = pd.concat([quarterly_volatility_df, quarter_country_df])
        
quarterly_volatility_df['Report Date'] = quarterly_volatility_df['Report Date'] + pd.offsets.MonthEnd(0)
```

```python
quarterly_volatility_df = quarterly_volatility_df.dropna().reset_index(drop=True)
```

```python
quarterly_volatility_df['Indicator Name'] = 'MSCI Index Volatlity'
quarterly_volatility_df['Indicator Code'] = 'CM.MKT.MSCI.VOLAT'
quarterly_volatility_df['Source'] = 'Calculated by author'
```

```python
quarterly_volatility_df = quarterly_volatility_df[['Country Name'
                                                   , 'Country Code'

                                                   , 'Indicator Name'
                                                   , 'Indicator Code'
                                                   
                                                   , 'Report Date'
                                                   , 'Value'
                                                   
                                                   , 'Source']]
```

# Quartelry exchange rates

```python
# loading up the data
daily_FOREX_df = pd.read_excel(r'data\prev_calc\daily\daily_FOREX.xlsx')
```

```python
quarterly_FOREX = pd.DataFrame()

for country in daily_FOREX_df['Country Name'].unique():
    daily_FOREX_country_temp = daily_FOREX_df[daily_FOREX_df['Country Name'] == country]
    
    for year in daily_FOREX_country_temp['Report Date'].dt.year.unique():
        daily_FOREX_year_temp = daily_FOREX_country_temp[daily_FOREX_country_temp['Report Date'].dt.year == year]
        
        for quarter in daily_FOREX_year_temp['Report Date'].dt.quarter.unique():
            daily_FOREX_quarter_temp = daily_FOREX_year_temp[daily_FOREX_year_temp['Report Date'].dt.quarter == quarter]
            
            FOREX_quarter = daily_FOREX_quarter_temp['Value'].mean()
            
            daily_FOREX_quarter_temp = daily_FOREX_quarter_temp[daily_FOREX_quarter_temp['Report Date'] == max(daily_FOREX_quarter_temp['Report Date'])]
            daily_FOREX_quarter_temp['Value'] = FOREX_quarter
            quarterly_FOREX = pd.concat([quarterly_FOREX, daily_FOREX_quarter_temp])

quarterly_FOREX['Report Date'] = quarterly_FOREX['Report Date'] + pd.offsets.MonthEnd(0)
```

### Current US$

```python
# loading up the data
usa_gdp_quarterly_curr_USD_df = pd.read_excel(r'data\raw\usa quarterly deflator\Table 1.1.5. Gross Domestic Product.xlsx')

usa_gdp_quarterly_curr_USD_df = usa_gdp_quarterly_curr_USD_df.iloc[4:].T

usa_gdp_quarterly_curr_USD_df[4] = usa_gdp_quarterly_curr_USD_df[4].ffill()

usa_gdp_quarterly_curr_USD_df.columns = usa_gdp_quarterly_curr_USD_df.iloc[1]

usa_gdp_quarterly_curr_USD_df = usa_gdp_quarterly_curr_USD_df.iloc[2:]

usa_gdp_quarterly_curr_USD_df.columns.values[0] = 'Report year'
usa_gdp_quarterly_curr_USD_df.columns.values[1] = 'Report quarter'

usa_gdp_quarterly_curr_USD_df['Report Date'] = pd.to_datetime(usa_gdp_quarterly_curr_USD_df['Report year'].astype(str) + '.' + \
                                                       (usa_gdp_quarterly_curr_USD_df['Report quarter'].apply(lambda x: 1 if 'Q1' in x \
                                                                                                   else 2 if 'Q2' in x \
                                                                                                   else 3 if 'Q3' in x \
                                                                                                   else 4 if 'Q4' in x \
                                                                                                   else None) * 3).astype(str) + '.01') + pd.offsets.MonthEnd(0)

usa_gdp_quarterly_curr_USD_df = pd.melt(usa_gdp_quarterly_curr_USD_df, id_vars=list(usa_gdp_quarterly_curr_USD_df.columns[-1:]), value_vars=list(usa_gdp_quarterly_curr_USD_df.columns[2:-1]))

usa_gdp_quarterly_curr_USD_df = usa_gdp_quarterly_curr_USD_df.rename(columns={'Unnamed: 1' : 'Indicator Name'
                                                                               , 'value' : 'Value'})

usa_gdp_quarterly_curr_USD_df['Indicator Name'] = usa_gdp_quarterly_curr_USD_df['Indicator Name'].str.strip() + ' (current US$)'

usa_gdp_quarterly_curr_USD_df = usa_gdp_quarterly_curr_USD_df.dropna()
```

```python
usa_gdp_quarterly_curr_USD_df = usa_gdp_quarterly_curr_USD_df[usa_gdp_quarterly_curr_USD_df['Indicator Name']=='Gross domestic product (current US$)']
```

```python
INDIA_gdp_quarterly_curr_LCU_1996_2010_df = pd.melt(INDIA_gdp_quarterly_curr_LCU_1996_2010_df, id_vars=list(INDIA_gdp_quarterly_curr_LCU_1996_2010_df.columns[-1:]), value_vars=list(INDIA_gdp_quarterly_curr_LCU_1996_2010_df.columns[2:-1]))
```

### Stocks traded, total value

```python
monthly_traded = monthly_WFE[monthly_WFE['Indicator Code']=='CM.MKT.TRAD.CD']
```

```python
monthly_traded = monthly_traded[monthly_traded['Exchange Code'].isin(['NSE+BSE','SSE+SZSE'])]
```

```python
quarterly_curr_traded = pd.DataFrame()

for country in monthly_traded['Country Name'].unique():
    monthly_traded_country_temp = monthly_traded[monthly_traded['Country Name'] == country]
    
    for year in monthly_traded_country_temp['Report Date'].dt.year.unique():
        monthly_traded_country_year_temp = monthly_traded_country_temp[monthly_traded_country_temp['Report Date'].dt.year == year]
        
        for quarter in monthly_traded_country_year_temp['Report Date'].dt.quarter.unique():
            traded_quarter_temp = monthly_traded_country_year_temp[monthly_traded_country_year_temp['Report Date'].dt.quarter == quarter]
            
            traded_quarter = traded_quarter_temp['Value'].sum()
            
            traded_quarter_temp = traded_quarter_temp[traded_quarter_temp['Report Date'] == max(traded_quarter_temp['Report Date'])]
            traded_quarter_temp['Value'] = traded_quarter
            quarterly_curr_traded = pd.concat([quarterly_curr_traded, traded_quarter_temp])

quarterly_curr_traded['Indicator Name'] = 'Stocks traded, total value (current LCU)'
quarterly_curr_traded['Indicator Code'] = 'CM.MKT.TRAD.CN'
quarterly_curr_traded['Source'] = 'Calculated by author'

quarterly_curr_traded['Report Date'] = quarterly_curr_traded['Report Date'] + pd.offsets.MonthEnd(0)
```
