# Transforming World Development Indicators (WDI)

## 2026 version

```python
# loading up the data
WDI_2026_02_df = pd.read_excel(r'data\WDI_2026_02.xlsx')
```

```python
# taking the necescary data and turning into tablular data
WDI_2026_02_df = pd.melt(WDI_2026_02_df, id_vars=list(WDI_2026_02_df.columns)[:4], value_vars=list(WDI_2026_02_df.columns)[4:])
```

```python
# renaming
WDI_2026_02_df.rename(columns={'variable': 'Year'
                          , 'value': 'Value'
                         }
                 , inplace=True)
```

```python
# making sure the data is an number
WDI_2026_02_df['Value'] = pd.to_numeric(WDI_2026_02_df['Value'], errors='coerce')
```

```python
WDI_2026_02_df.dropna(inplace=True)
```

```python
WDI_2026_02_df['Year'] = WDI_2026_02_df['Year'].str[:4]
```

```python
# taking only the necescary years
WDI_2026_02_df['Year'] = WDI_2026_02_df['Year'].astype(int)
WDI_2026_02_df = WDI_2026_02_df[WDI_2026_02_df['Year']>1989]
```

```python
WDI_2026_02_df['Source']= 'World Bank 2026'
```

## Combining the versions and exporting

```python
WDI_df = pd.concat([WDI_2014_12_df, WDI_2026_02_df, WDI_2025_06_df, india_finance_df]).drop_duplicates(subset = ['Country Name','Country Code','Series Name','Series Code','Year'], keep='last').reset_index(drop=True)
```

```python
# loading up data from a previous study
WDI_df.to_excel(r'formated_data\WDI_full_collected.xlsx', index=False)
```

# Calcualting absolute values

```python
WDI_df = WDI_df.pivot_table(index=['Country Name', 'Year'],
                            columns=['Series Name'],
                            values=['Value'],
                            aggfunc='sum')
```

```python
# fixing column names
WDI_df.columns = WDI_df.columns.droplevel(level=0)
```

```python
WDI_df = WDI_df.reset_index(drop=False)
```

```python
dependant_variable_list = ['GDP (constant 2015 US$)'
                           , 'GDP (current US$)'
                          ]
```

```python
independant_variable_list = ['Domestic credit to private sector (% of GDP)'
                             , 'Exports of goods and services (% of GDP)'
                             , 'Market capitalization of listed domestic companies (% of GDP)'
                            ]
```

```python
for dependant_variable in dependant_variable_list:
    # print(dependant_variable)
    unit_type = dependant_variable[4:]
    for independant_variable in independant_variable_list:
        # print(independant_variable)
        WDI_df[f'''{independant_variable.replace('(% of GDP)', unit_type)}'''] = WDI_df[independant_variable] * WDI_df[dependant_variable] / 100
```

# Exporting

```python
WDI_df = pd.melt(WDI_df, id_vars=list(WDI_df.columns)[:2], value_vars=list(WDI_df.columns)[2:])
```

```python
WDI_df.dropna(inplace=True)
```

```python
series_dict = {'Domestic credit to private sector (% of GDP)' : 'FS.AST.PRVT.GD.ZS'
               , 'Exports of goods and services (% of GDP)' : 'NE.EXP.GNFS.ZS'
               , 'Market capitalization of listed domestic companies (% of GDP)' : 'CM.MKT.LCAP.GD.ZS'
               
               , 'Domestic credit to private sector (constant 2015 US$)' : 'FS.AST.PRVT.GD.KD'
               , 'Exports of goods and services (constant 2015 US$)' : 'NE.EXP.GNFS.KD'
               , 'Market capitalization of listed domestic companies (constant 2015 US$)' : 'CM.MKT.LCAP.GD.KD'
               
               , 'Domestic credit to private sector (current US$)' : 'FS.AST.PRVT.GD.CD'
               , 'Exports of goods and services (current US$)' : 'NE.EXP.GNFS.CD'
               , 'Market capitalization of listed domestic companies (current US$)' : 'CM.MKT.LCAP.GD.CD'
               
               , 'GDP (constant 2015 US$)' : 'NY.GDP.MKTP.KD'
               , 'GDP (current US$)' : 'NY.GDP.MKTP.CD'
              }
```

```python
country_dict = {'India' : 'IND'
                , 'China' : 'CHN'
               }
```

```python
WDI_df['Country Code'] = WDI_df['Country Name'].apply(lambda x: country_dict.get(x))
WDI_df['Series Code'] = WDI_df['Series Name'].apply(lambda x: series_dict.get(x))
```

```python
# renaming
WDI_df.rename(columns={'value': 'Value'}, inplace=True)
```

```python
WDI_df = WDI_df[['Country Name'
                 , 'Country Code'
                 , 'Series Name'
                 , 'Series Code'
                 , 'Year'
                 , 'Value'
                ]]
```
