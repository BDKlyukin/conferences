# Exhange rates - IMF

```python
# loading up the data
IMF_forex_daily_df = pd.read_excel(r'data\forex_data\IMF_curr_exchange_rates.xlsx')
```

```python
# renaming columns
IMF_forex_daily_df.rename(columns={IMF_forex_daily_df.columns.tolist()[0]:'Report Date'
                                   , IMF_forex_daily_df.columns.tolist()[1]:'Chinese yuan (CNY)'
                                   , IMF_forex_daily_df.columns.tolist()[2]:'Indian rupee (INR)'
                                  }
                          , inplace = True)
```

```python
IMF_forex_daily_df = IMF_forex_daily_df.iloc[2:7885].reset_index(drop=True)
```

```python
# making sure that the report date columns is a actualy in the proper date format
IMF_forex_daily_df['Report Date'] = pd.to_datetime(IMF_forex_daily_df['Report Date'])
```

## China

```python
# taking only china
china_IMF_forex_daily_df = IMF_forex_daily_df[['Report Date', 'Chinese yuan (CNY)']].dropna(subset=['Chinese yuan (CNY)'])

# adding the country meta data
china_IMF_forex_daily_df['Country Name'] = 'China'
china_IMF_forex_daily_df['Country Code'] = 'CHN'

# renaming
china_IMF_forex_daily_df.rename(columns={'Chinese yuan (CNY)': 'Value'
                                        }
                                , inplace = True)
```

## India

```python
# taking only india
india_IMF_forex_daily_df = IMF_forex_daily_df[['Report Date', 'Indian rupee (INR)']].dropna(subset=['Indian rupee (INR)'])

# adding the country meta data
india_IMF_forex_daily_df['Country Name'] = 'India'
india_IMF_forex_daily_df['Country Code'] = 'IND'

# renaming
india_IMF_forex_daily_df.rename(columns={'Indian rupee (INR)': 'Value'
                                        }
                                , inplace = True)
```

## Together

```python
# stitching them newer raw data with hand filled data
IMF_forex_daily_df = pd.concat([india_IMF_forex_daily_df, china_IMF_forex_daily_df])
```

```python
# more meta data
IMF_forex_daily_df['Indicator Name'] = 'IMF exchange rate (LCU per US$, period average)'
IMF_forex_daily_df['Indicator Code'] = 'PA.NUS.FCRF'
IMF_forex_daily_df['Source'] = 'IMF'
```

```python
# taking the main columns
IMF_forex_daily_df = IMF_forex_daily_df[['Country Name'
                                         , 'Country Code'
                                         
                                         , 'Indicator Name'
                                         , 'Indicator Code'
                                         
                                         , 'Report Date'
                                         , 'Value'
                                         
                                         , 'Source'
                                        ]]
```

## Exporting and formatting for further use

```python
IMF_forex_daily_df.to_excel(r'formated_data\_IMF_forex_daily.xlsx', index=False)
```

```python
IMF_forex_2015_df = IMF_forex_daily_df[IMF_forex_daily_df['Report Date'].dt.year==2015]
```

```python
IMF_forex_2015_df['Year'] = IMF_forex_2015_df['Report Date'].dt.year
```

```python
IMF_forex_2015_df = IMF_forex_2015_df.groupby(['Country Name','Country Code','Indicator Name','Indicator Code','Source','Year']).mean().reset_index(drop=False)
```

```python
IMF_forex_2015_df['Report Date'] = pd.to_datetime(IMF_forex_2015_df['Year'], format='%Y') + pd.offsets.YearEnd(0)
```

# Quarterly deflator - OECD via FRED

```python
# loading up the data
FRED_Q_INF_CHINA_CURR_LCU = pd.read_excel(r'data\deflator_data\FRED_Q_INF_CHINA_CURR_LCU.xlsx', 'Quarterly')
FRED_Q_INF_CHINA_CURR_LCU['observation_date'] = FRED_Q_INF_CHINA_CURR_LCU['observation_date']  + pd.offsets.MonthEnd(3)

FRED_Q_INF_CHINA_CURR_LCU['Country Name'] = 'China'
FRED_Q_INF_CHINA_CURR_LCU['Country Code'] = 'CHN'

FRED_Q_INF_CHINA_CURR_LCU = FRED_Q_INF_CHINA_CURR_LCU.rename(columns={'observation_date' : 'Report Date'
                                                                      , 'CHNCPIALLQINMEI' : 'Value'})
```

```python
# loading up the data
FRED_Q_INF_INDIA_CURR_LCU = pd.read_excel(r'data\deflator_data\FRED_Q_INF_INDIA_CURR_LCU.xlsx', 'Quarterly')
FRED_Q_INF_INDIA_CURR_LCU['observation_date'] = FRED_Q_INF_INDIA_CURR_LCU['observation_date']  + pd.offsets.MonthEnd(3)

FRED_Q_INF_INDIA_CURR_LCU['Country Name'] = 'India'
FRED_Q_INF_INDIA_CURR_LCU['Country Code'] = 'IND'

FRED_Q_INF_INDIA_CURR_LCU = FRED_Q_INF_INDIA_CURR_LCU.rename(columns={'observation_date' : 'Report Date'
                                                                      , 'INDCPIALLQINMEI' : 'Value'})
```

```python
deflator_lcu_df = pd.concat([FRED_Q_INF_CHINA_CURR_LCU, FRED_Q_INF_INDIA_CURR_LCU])
```

```python
deflator_lcu_df['Value'] = deflator_lcu_df['Value']/100

deflator_lcu_df['Source'] = 'Federal Reserve Bank of St. Louis'

deflator_lcu_df['Indicator Code']	= 'NY.GDP.DEFL.KN.ZG'
deflator_lcu_df['Indicator Name'] = 'GDP deflator (2015 LCU)'
```

```python
deflator_lcu_df.to_excel(r'formated_data\_deflator_lcu.xlsx', index=False)
```

### India

#### Series 1999-00

```python
# loading up the data
INDIA_gdp_quarterly_curr_LCU_1999_2000_df = pd.read_excel(r'data\econ_data\INDIA_GVA_QUARTERLY.xlsx', 'NAS� 1999-2000')
```

```python
column_names = INDIA_gdp_quarterly_curr_LCU_1999_2000_df.iloc[4][1:].str.strip()
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df = INDIA_gdp_quarterly_curr_LCU_1999_2000_df.iloc[5:-2].T.iloc[1:].T
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df.columns = column_names
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'] = INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'].ffill()
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'] = INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'].str.strip()
```

```python
date_corrector = lambda x: str(x['Year/ Industry'][:2]) + str(x['Year/ Industry'][-2:]) if x['Quarter']=='Q4' else str(x['Year/ Industry'][:4])
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'] = INDIA_gdp_quarterly_curr_LCU_1999_2000_df.apply(lambda x: date_corrector(x), axis=1)
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'] = INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'].apply(lambda x: '2000' if x == '1900' else x)
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Quarter'] = INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Quarter'].apply(lambda x : 1 if int(x[-1]) == 4 else int(x[-1])+1)*3
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Report Date'] = pd.to_datetime(INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Year/ Industry'].astype(str) + '.' + INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Quarter'].astype(str) + '.01') + pd.offsets.MonthEnd(0)
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df = pd.melt(INDIA_gdp_quarterly_curr_LCU_1999_2000_df, id_vars=list(INDIA_gdp_quarterly_curr_LCU_1999_2000_df.columns[-1:]), value_vars=list(INDIA_gdp_quarterly_curr_LCU_1999_2000_df.columns[2:-1]))
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df['Source'] = 'MOSPI Series 1999-00'
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df.columns.values[1] = 'Sector'
INDIA_gdp_quarterly_curr_LCU_1999_2000_df.columns.values[2] = 'Value (Rupees Crores)'
```

```python
INDIA_gdp_quarterly_curr_LCU_1999_2000_df.to_excel(r'formated_data\_INDIA_gdp_quarterly_series_1999_2000.xlsx', index=False)
```

```python
# calculating the nominal values
WFE_original_df['Value'] = WFE_original_df['Value'] * WFE_original_df['Nominal']
```

```python
# translating the english months to a number
WFE_original_df['Month'] = WFE_original_df['Month'].apply(lambda x: month_to_number_dict.get(x))
```

```python
# building a date
WFE_original_df['Report Date'] = pd.to_datetime(WFE_original_df['Year'].astype(str) + '.' + WFE_original_df['Month'].astype(str) + '.' + '01') + pd.DateOffset(months=1) - timedelta(days=1)
```

```python
# keeping the main columns
WFE_original_df = WFE_original_df[['Indicator Name'
                         , 'ExchangeName'
                         , 'Report Date'
                         , 'Value'
                        ]]
```

```python
# formatting the indicator names
WFE_original_df['Indicator Name'] = WFE_original_df['Indicator Name'].str.replace('Total Equity Market - ', '')
```

#### Calculating the value traded (it is split over three different categories of trading)

```python
# taking the rows which don't contain data on trading
WFE_not_value_traded_df = WFE_original_df[~WFE_original_df['Indicator Name'].str.contains('trade')]
```

```python
# taking that do and aggregating them
WFE_value_traded_df = WFE_original_df[WFE_original_df['Indicator Name'].str.contains('trade')].groupby(by=['Report Date', 'ExchangeName']).sum()
```

```python
# formating
WFE_value_traded_df = WFE_value_traded_df.reset_index(drop=False)
WFE_value_traded_df['Indicator Name'] = 'Stocks traded, total value (current LCU)'
```

```python
# stitching them back together
WFE_original_df = pd.concat([WFE_not_value_traded_df, WFE_value_traded_df])
```

## Volatility (quarterly)

```python
MSCI_index_quarter_volat_df = pd.DataFrame()

for country in MSCI_index_daily_value_df['Country Name'].unique():
    MSCI_index_country_temp = MSCI_index_daily_value_df[MSCI_index_daily_value_df['Country Name'] == country]
    
    for year in MSCI_index_country_temp['Report Date'].dt.year.unique():
        MSCI_index_year_temp = MSCI_index_country_temp[MSCI_index_country_temp['Report Date'].dt.year == year]
        
        for quarter in MSCI_index_year_temp['Report Date'].dt.quarter.unique():
            MSCI_index_quarter_temp = MSCI_index_year_temp[MSCI_index_year_temp['Report Date'].dt.quarter == quarter]
            
            MSCI_index_volatility = ((MSCI_index_quarter_temp['Value'] / MSCI_index_quarter_temp['Value'].shift(1)).dropna()-1).std() * np.sqrt(252)
            
            MSCI_index_quarter_temp = MSCI_index_quarter_temp[MSCI_index_quarter_temp['Report Date'] == max(MSCI_index_quarter_temp['Report Date'])]
            MSCI_index_quarter_temp['Value'] = MSCI_index_volatility
            MSCI_index_quarter_volat_df = pd.concat([MSCI_index_quarter_volat_df, MSCI_index_quarter_temp])

MSCI_index_quarter_volat_df['Indicator Name'] = 'MSCI Index Volatlity'
MSCI_index_quarter_volat_df['Indicator Code'] = 'CM.MKT.MSCI.INDEX.VOLAT'
MSCI_index_quarter_volat_df['Source'] = 'Calculated by author'

MSCI_index_quarter_volat_df['Report Date'] = MSCI_index_quarter_volat_df['Report Date'] + pd.offsets.MonthEnd(0)
```

```python
MSCI_index_quarter_volat_df.to_excel(r'formated_data/_MSCI_index_quarter_volat.xlsx', index=False)
```

```python
# making sure that the report date columns is a actualy in the proper date format
MSCI_pe_monthly_value_df['Report Date'] = pd.to_datetime(MSCI_pe_monthly_value_df['Report Date'])
```

### China

```python
# taking only china
china_MSCI_pe_monthly_value_df = MSCI_pe_monthly_value_df[['Report Date', 'China']].dropna(subset=['China'])

# adding the country meta data
china_MSCI_pe_monthly_value_df['Country Name'] = 'China'
china_MSCI_pe_monthly_value_df['Country Code'] = 'CHN'

# renaming
china_MSCI_pe_monthly_value_df.rename(columns={'China': 'Value'
                                }
                        , inplace = True)
```

### India

```python
# taking only china
india_MSCI_pe_monthly_value_df = MSCI_pe_monthly_value_df[['Report Date', 'India']].dropna(subset=['India'])

# adding the country meta data
india_MSCI_pe_monthly_value_df['Country Name'] = 'India'
india_MSCI_pe_monthly_value_df['Country Code'] = 'IND'

# renaming
india_MSCI_pe_monthly_value_df.rename(columns={'India': 'Value'
                                }
                        , inplace = True)
```

### Together

```python
# stitching the countries' data back together
MSCI_pe_monthly_value_df = pd.concat([india_MSCI_pe_monthly_value_df, china_MSCI_pe_monthly_value_df])
```
