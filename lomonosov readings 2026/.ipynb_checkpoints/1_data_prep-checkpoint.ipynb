```python
# taking only india
india_IMF_forex_daily_df = IMF_forex_daily_df[['Report Date', 'Indian rupee (INR)']].dropna(subset=['Indian rupee (INR)'])

# adding the country meta data
india_IMF_forex_daily_df['Country Name'] = 'India'
india_IMF_forex_daily_df['Country Code'] = 'IND'

# renaming
india_IMF_forex_daily_df.rename(columns={'Indian rupee (INR)': 'Value'
                                        }
                                , inplace = True)
```

## Annual

### WDI

```python
# loading up the data
WDI_forex_annual_df = pd.read_excel(r'data\raw\WDI_2025_07_FOREX.xlsx')
```

```python
# dropping the comments
WDI_forex_annual_df = WDI_forex_annual_df.dropna(subset=['Country Code'])

WDI_forex_annual_df = pd.melt(WDI_forex_annual_df, id_vars=list(WDI_forex_annual_df.columns)[:6], value_vars=list(WDI_forex_annual_df.columns)[6:])

WDI_forex_annual_df = WDI_forex_annual_df[['Country Name'
                                           , 'Country Code'
                                           , 'Series Name'
                                           , 'Series Code'
                                           , 'variable'
                                           , 'value'
                                          ]]

# renaming
WDI_forex_annual_df.rename(columns={'variable': 'Year'
                                    , 'value': 'Value'
                                    , 'Series Name': 'Indicator Name'
                                    , 'Series Code': 'Indicator Code'
                                   }
                           , inplace=True)

WDI_forex_annual_df['Year'] = WDI_forex_annual_df['Year'].str.split('['
                                                          , expand=True)[0]
WDI_forex_annual_df['Year'] = WDI_forex_annual_df['Year'].astype(int)

# making sure the data is an int
WDI_forex_annual_df['Value'] = pd.to_numeric(WDI_forex_annual_df['Value'], errors='coerce')

WDI_forex_annual_df['Source']= 'World Bank 2025 07'
```

### IMF + Investing

```python
forex_annual_df = pd.DataFrame()

for country in FOREX_daily_df['Country Name'].unique():
    forex_country_temp = FOREX_daily_df[FOREX_daily_df['Country Name'] == country]
    for year in forex_country_temp['Report Date'].dt.year.unique():
        forex_year_temp = forex_country_temp[forex_country_temp['Report Date'].dt.year == year]
        
        forex_value = forex_year_temp['Value'].mean()
        
        forex_year_temp = forex_year_temp[forex_year_temp['Report Date'] == max(forex_year_temp['Report Date'])]
        
        forex_year_temp['Value'] = forex_value
        forex_year_temp['Source'] = 'IMF and Investing.com'
        forex_year_temp['Report Date'] = forex_year_temp['Report Date'].dt.year
        forex_year_temp.rename(columns={'Report Date': 'Year'
                                   }
                           , inplace=True)
        
        forex_annual_df = pd.concat([forex_annual_df, forex_year_temp])
```

### Together

```python
forex_annual_df = pd.concat([forex_annual_df, WDI_forex_annual_df])
forex_annual_df = forex_annual_df[forex_annual_df['Year']>1990]
forex_annual_df = forex_annual_df.sort_values(by=['Year','Country Name']).reset_index(drop=True)
forex_annual_df.to_excel(r'transformed_data/annual_FOREX.xlsx', index=False)
```

#### 1996-2010

```python
# loading up the data
INDIA_gdp_quarterly_curr_LCU_1996_2010_df = pd.read_excel(r'data\raw\INDIA_GDP_QUARTERLY_CURR.xlsx', 'NAS � 1999-2000')

column_names = INDIA_gdp_quarterly_curr_LCU_1996_2010_df.iloc[4][1:].str.strip()

INDIA_gdp_quarterly_curr_LCU_1996_2010_df = INDIA_gdp_quarterly_curr_LCU_1996_2010_df.iloc[5:-2].T.iloc[1:].T

INDIA_gdp_quarterly_curr_LCU_1996_2010_df.columns = column_names

INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'] = INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'].ffill()

INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'] = INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'].str.strip()

date_corrector = lambda x: str(x['Industry/ Year'][:2]) + str(x['Industry/ Year'][-2:]) if x['Quarter']=='Q4' else str(x['Industry/ Year'][:4])
INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'] = INDIA_gdp_quarterly_curr_LCU_1996_2010_df.apply(lambda x: date_corrector(x), axis=1)
INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'] = INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'].apply(lambda x: '2000' if x == '1900' else x)
INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Quarter'] = INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Quarter'].apply(lambda x : 1 if int(x[-1]) == 4 else int(x[-1])+1)*3

INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Report Date'] = pd.to_datetime(INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Industry/ Year'].astype(str) + '.' + INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Quarter'].astype(str) + '.01') + pd.offsets.MonthEnd(0)
```

```python
INDIA_gdp_quarterly_curr_LCU_1996_2010_df = pd.melt(INDIA_gdp_quarterly_curr_LCU_1996_2010_df, id_vars=list(INDIA_gdp_quarterly_curr_LCU_1996_2010_df.columns[-1:]), value_vars=list(INDIA_gdp_quarterly_curr_LCU_1996_2010_df.columns[2:-1]))
```

```python
INDIA_gdp_quarterly_curr_LCU_1996_2010_df['Source'] = 'Series 1999-00'
```

```python
INDIA_gdp_quarterly_curr_LCU_1996_2010_df = INDIA_gdp_quarterly_curr_LCU_1996_2010_df[INDIA_gdp_quarterly_curr_LCU_1996_2010_df[4]=='Gross Domestic Product']
```

## Current US$

### quarterly FOREX

```python
FOREX_quarterly_df = pd.DataFrame()

for country in FOREX_daily_df['Country Name'].unique():
    daily_FOREX_country_temp = FOREX_daily_df[FOREX_daily_df['Country Name'] == country]
    
    for year in daily_FOREX_country_temp['Report Date'].dt.year.unique():
        daily_FOREX_year_temp = daily_FOREX_country_temp[daily_FOREX_country_temp['Report Date'].dt.year == year]
        
        for quarter in daily_FOREX_year_temp['Report Date'].dt.quarter.unique():
            daily_FOREX_quarter_temp = daily_FOREX_year_temp[daily_FOREX_year_temp['Report Date'].dt.quarter == quarter]
            
            FOREX_quarter = daily_FOREX_quarter_temp['Value'].mean()
            
            daily_FOREX_quarter_temp = daily_FOREX_quarter_temp[daily_FOREX_quarter_temp['Report Date'] == max(daily_FOREX_quarter_temp['Report Date'])]
            daily_FOREX_quarter_temp['Value'] = FOREX_quarter
            FOREX_quarterly_df = pd.concat([FOREX_quarterly_df, daily_FOREX_quarter_temp])

FOREX_quarterly_df['Report Date'] = FOREX_quarterly_df['Report Date'] + pd.offsets.MonthEnd(0)
```

### Calculating

```python
gdp_quarterly_curr_USD_df = pd.merge(gdp_quarterly_curr_LCU_df, FOREX_quarterly_df, on =['Report Date', 'Country Name', 'Country Code'], how='left')
```

```python
gdp_quarterly_curr_USD_df = gdp_quarterly_curr_USD_df.drop_duplicates(subset=['Report Date', 'Indicator Name_x', 'Country Name', 'Country Code', 'Indicator Code_x'], keep='last')#.sort_values(by=['Report Date']).reset_index(drop = True)
```

```python
gdp_quarterly_curr_USD_df['Value'] = gdp_quarterly_curr_USD_df['Value_x'] / gdp_quarterly_curr_USD_df['Value_y']

gdp_quarterly_curr_USD_df['Source'] = 'Calculated by author'

gdp_quarterly_curr_USD_df['Indicator Code']	= 'NY.GDP.MKTP.CD'
gdp_quarterly_curr_USD_df['Indicator Name'] = 'GDP (current US$)'

gdp_quarterly_curr_USD_df = gdp_quarterly_curr_USD_df[['Country Name'
                                                       , 'Country Code'
                                                       , 'Indicator Name'
                                                       , 'Indicator Code'
                                                       , 'Report Date'
                                                       , 'Value'
                                                       , 'Source'
                                                      ]]
```

## World Federation of Exchanges (WFE) - Raw data

```python
# loading up the data
WFE_raw_df = pd.read_excel(r'data\raw\WFE_report_10_06_2025_CLC.xlsx')
```

```python
# dropping the rows with no data
WFE_raw_df.dropna(subset=['Value'], inplace=True)
```

```python
# calculating the nominal values
WFE_raw_df['Value'] = WFE_raw_df['Value'] * WFE_raw_df['Nominal']
```

```python
# translating the english months to a number
WFE_raw_df['Month'] = WFE_raw_df['Month'].apply(lambda x: month_to_number_dict.get(x))
```

```python
# building a date
WFE_raw_df['Report Date'] = pd.to_datetime(WFE_raw_df['Year'].astype(str) + '.' + WFE_raw_df['Month'].astype(str) + '.' + '01') + pd.DateOffset(months=1) - timedelta(days=1)
```

```python
# keeping the main columns
WFE_raw_df = WFE_raw_df[['Indicator Name'
                         , 'ExchangeName'
                         , 'Report Date'
                         , 'Value'
                        ]]
```

```python
# formatting the indicator names
WFE_raw_df['Indicator Name'] = WFE_raw_df['Indicator Name'].str.replace('Total Equity Market - ', '')
```

### Calculating the value traded (it is split over three different categories of trading)

```python
# taking the rows which don't contain data on trading
WFE_not_value_traded_df = WFE_raw_df[~WFE_raw_df['Indicator Name'].str.contains('trade')]
```

```python
# taking that do and aggregating them
WFE_value_traded_df = WFE_raw_df[WFE_raw_df['Indicator Name'].str.contains('trade')].groupby(by=['Report Date', 'ExchangeName']).sum()
```

```python
# formating
WFE_value_traded_df = WFE_value_traded_df.reset_index(drop=False)
WFE_value_traded_df['Indicator Name'] = 'Stocks traded, total value'
```

```python
# stitching them back together
WFE_raw_df = pd.concat([WFE_not_value_traded_df, WFE_value_traded_df])
```

### Final formating for WFE data

```python
# working on proper nomenculture of the indicators
WFE_raw_df['Indicator Name'] = WFE_raw_df['Indicator Name'].str.replace('Market Capitalisation', 'Market capitalization of listed domestic companies')
WFE_raw_df['Indicator Name'] = WFE_raw_df['Indicator Name'].str.replace('Number of listed companies (Total)', 'Listed domestic companies, total')
```

```python
# fixing the column name
WFE_raw_df.rename(columns={'ExchangeName':'Exchange Name'
                      }
              , inplace = True)
```

```python
# assiging meta data
WFE_raw_df['Source'] = 'World Federation of Exchanges'

WFE_raw_df['Country Name'] = WFE_raw_df['Exchange Name'].apply(lambda x: 'India' if 'India' in x 
                                                       else 'India' if 'Bombay' in x 
                                                       else 'China')

WFE_raw_df['Country Code'] = WFE_raw_df['Country Name'].apply(lambda x: 'IND' if x=='India' 
                                                      else 'CHN')

WFE_raw_df['Indicator Code'] = WFE_raw_df['Indicator Name'].apply(lambda x: 'CM.MKT.LCAP.CN' if x == 'Market capitalization of listed domestic companies'
                                                          else 'CM.MKT.TRAD.CN' if x == 'Stocks traded, total value'
                                                          else 'CM.MKT.LDOM.NO')

WFE_raw_df['Exchange Code'] = WFE_raw_df['Exchange Name'].apply(lambda x: 'NSE' if x=='National Stock Exchange of India'
                                                        else 'BSE' if x == 'BSE India Limited'
                                                        else 'SZSE' if x == 'Shenzhen Stock Exchange'
                                                        else 'SSE')
```

```python
# taking the main columns
WFE_raw_df = WFE_raw_df[['Country Name'
                         , 'Country Code'
                         
                         , 'Exchange Name'
                         , 'Exchange Code'
                         
                         , 'Indicator Name'
                         , 'Indicator Code'
                         
                         , 'Report Date'
                         , 'Value'
                         
                         , 'Source'
                        ]]
```

## Volatility (quarterly)

```python
MSCI_index_quarter_volat_df = pd.DataFrame()

for country in MSCI_index_daily_value_df['Country Name'].unique():
    MSCI_index_country_temp = MSCI_index_daily_value_df[MSCI_index_daily_value_df['Country Name'] == country]
    
    for year in MSCI_index_country_temp['Report Date'].dt.year.unique():
        MSCI_index_year_temp = MSCI_index_country_temp[MSCI_index_country_temp['Report Date'].dt.year == year]
        
        for quarter in MSCI_index_year_temp['Report Date'].dt.quarter.unique():
            MSCI_index_quarter_temp = MSCI_index_year_temp[MSCI_index_year_temp['Report Date'].dt.quarter == quarter]
            
            MSCI_index_volatility = ((MSCI_index_quarter_temp['Value'] / MSCI_index_quarter_temp['Value'].shift(1)).dropna()-1).std() * np.sqrt(252)
            
            MSCI_index_quarter_temp = MSCI_index_quarter_temp[MSCI_index_quarter_temp['Report Date'] == max(MSCI_index_quarter_temp['Report Date'])]
            MSCI_index_quarter_temp['Value'] = MSCI_index_volatility
            MSCI_index_quarter_volat_df = pd.concat([MSCI_index_quarter_volat_df, MSCI_index_quarter_temp])

MSCI_index_quarter_volat_df['Indicator Name'] = 'MSCI Index Volatlity'
MSCI_index_quarter_volat_df['Indicator Code'] = 'CM.MKT.MSCI.INDEX.VOLAT'
MSCI_index_quarter_volat_df['Source'] = 'Calculated by author'

MSCI_index_quarter_volat_df['Report Date'] = MSCI_index_quarter_volat_df['Report Date'] + pd.offsets.MonthEnd(0)
```

```python
MSCI_index_quarter_volat_df.to_excel(r'transformed_data/MSCI_index_quarter_volat.xlsx', index=False)
```
