```python
# loading up the data
data_df = pd.read_excel(r'model_data/quarterly_const_LCU.xlsx', index_col='Report Date' )

data_df = data_df.pivot_table(values='Value'
                                          , index=['Report Date']
                                          , columns='Indicator Name').reset_index(drop=False)

data_df.rename(columns={'Market capitalization of listed domestic companies (constant LCU)' : 'Капитализация'
                     , 'MSCI Index Volatlity' : 'Волатильность'
                     , 'MSCI PE Value' : 'Цена/прибыль'
                     , 'Stocks traded, total value (constant LCU)' : 'Объем торгов'
                     , 'GDP (constant LCU)' : 'ВВП'
                    }
            , inplace=True)
```

```python
data_df = data_df.set_index('Report Date')

data_df['Оборачиваемость'] = (data_df['Объем торгов'] / data_df['Капитализация'])

data_df = data_df.dropna()
```

```python
data_df
```

# Seasonal decomposition

```python
for column_name in data_df.columns:
    
    result = seasonal_decompose(data_df[column_name], model='additive', period=4)
    
    deseasoned_df = pd.DataFrame(result.observed - result.seasonal, columns=[column_name + ' (сезонность устранена) Classical'])
    
    data_df = pd.concat([data_df, deseasoned_df], axis=1)
```

```python
for column_name in data_df.columns[:6]:

    stl = STL(data_df[column_name], seasonal=5, low_pass=21, robust=True)
    result = stl.fit()

    deseasoned_df = pd.DataFrame(result.observed - result.seasonal, columns=[column_name + ' (сезонность устранена) STL'])
    
    data_df = pd.concat([data_df, deseasoned_df], axis=1)
```

```python
data_df.T.sort_index().T.to_excel(r'explore_data\_time_series_decomposition_quarterly.xlsx', index=True)
```

```python
data_df = data_df[['ВВП (сезонность устранена) STL'
                   , 'Капитализация'
                   , 'Объем торгов'
                   , 'Оборачиваемость'
                   , 'Волатильность'
                   , 'Цена/прибыль'
                  ]]

# data_df = data_df[data_df.columns[6:-6]]
```

```python
# buidling the df
stationarity_df = pd.DataFrame({
    'Parameter' : [],
    'ADF p-value (level data)': [],
    # 'DFGLS p-value': [],
    # 'PhillipsPerron p-value': [],
    'KPSS p-value (level data)': [],
    # 'ZivotAndrews p-value': []
})

for column in data_df.columns:

    adf_result = ADF(data_df[column]
                     , lags = 1
                     , trend = 'n'
                     # # , max_lags = 8
                     , method = 'aic'
                     ).pvalue

    dfgls_result = DFGLS(data_df[column]
                         , lags = 1
                         , trend = 'c'
                         # , max_lags = 8
                         , method = 'aic'
                        ).pvalue

    PhillipsPerron_result = PhillipsPerron(data_df[column]
                                           , lags = 1
                                           , trend = 'c'
                                           , test_type = 'tau'
                                          ).pvalue

    kpss_result = kpss(data_df[column].dropna()
                       , regression='c'
                       , nlags='auto'
                       , store=False)[1]

    ZivotAndrews_result = ZivotAndrews(data_df[column]
                                           , lags = 1
                                           , trend = 'c'
                                           , trim = 0.15
                                           # , max_lags = 8
                                           , method = 'aic'
                                          ).pvalue
    # buidling the df
    stationarity_temp_df = pd.DataFrame({
        'Parameter' : [column],
        'ADF p-value (level data)': [round(adf_result, 4)],
        # 'DFGLS p-value': [round(dfgls_result, 4)],
        # 'PhillipsPerron p-value': [round(PhillipsPerron_result, 4)],
        'KPSS p-value (level data)': [round(kpss_result, 4)],
        # 'ZivotAndrews p-value': [round(ZivotAndrews_result, 4)]
    })
    
    stationarity_df = pd.concat([stationarity_df, stationarity_temp_df])
```

```python
# diff_data_df = data_df.pct_change().dropna()
diff_data_df = data_df.diff().dropna()
```

```python
# buidling the df
integrated_stationarity_df = pd.DataFrame({
    'Parameter' : [],
    'ADF p-value (differenced data)': [],
    # 'DFGLS p-value': [],
    # 'PhillipsPerron p-value': [],
    'KPSS p-value (differenced data)': [],
    # 'ZivotAndrews p-value': []
})

for column in diff_data_df.columns:

    adf_result = ADF(diff_data_df[column]
                     , lags = 1
                     , trend = 'n'
                     # # , max_lags = 8
                     , method = 'aic'
                     ).pvalue

    dfgls_result = DFGLS(diff_data_df[column]
                         , lags = 1
                         , trend = 'c'
                         # , max_lags = 8
                         , method = 'aic'
                        ).pvalue

    PhillipsPerron_result = PhillipsPerron(diff_data_df[column]
                                           , lags = 1
                                           , trend = 'c'
                                           , test_type = 'tau'
                                          ).pvalue

    kpss_result = kpss(diff_data_df[column]
                       , regression='c'
                       , nlags='auto'
                       , store=False)[1]

    ZivotAndrews_result = ZivotAndrews(diff_data_df[column]
                                           , lags = 1
                                           , trend = 'c'
                                           , trim = 0.15
                                           # , max_lags = 8
                                           , method = 'aic'
                                          ).pvalue
    # buidling the df
    stationarity_temp_df = pd.DataFrame({
        'Parameter' : [column],
        'ADF p-value (differenced data)': [round(adf_result, 4)],
        # 'DFGLS p-value': [round(dfgls_result, 4)],
        # 'PhillipsPerron p-value': [round(PhillipsPerron_result, 4)],
        'KPSS p-value (differenced data)': [round(kpss_result, 4)],
        # 'ZivotAndrews p-value': [round(ZivotAndrews_result, 4)]
    })
    
    integrated_stationarity_df = pd.concat([integrated_stationarity_df, stationarity_temp_df])
```

```python
pd.merge(stationarity_df, integrated_stationarity_df, on='Parameter', how='left').to_excel(r'explore_data\_stationarity_analysis.xlsx', index=True)
```

## Cointegration test

```python
coint_test_df = pd.DataFrame()

for pairs in combinations(list(data_df.columns), 2):


    result = coint(data_df[pairs[0]], data_df[pairs[1]]
              , trend='ct')
    
    # creating a unit name to unit value dictionary
    coint_test = {'var_1': [pairs[0]]
                  , 'var_2': [pairs[1]]
                  , 'p_value': [result[1]]
                 }
    temp_coint_test_df = pd.DataFrame.from_dict(coint_test)
    coint_test_df = pd.concat([coint_test_df, temp_coint_test_df])
```

```python
# making a matrix of results
coint_test_df = coint_test_df.pivot_table(values='p_value'
                                     , index=['var_2']
                                     , columns=['var_1'])

# Sort the DataFrame by the number of non-null values
coint_test_df['non_null_count'] = coint_test_df.count(axis=1) # Calculate non-null counts per row
coint_test_df = coint_test_df.sort_values(by='non_null_count', ascending=False) # Sort in descending order
coint_test_df = coint_test_df.drop('non_null_count', axis=1)

# Sort the DataFrame by the number of non-null values
coint_test_df = coint_test_df.T
coint_test_df['non_null_count'] = coint_test_df.count(axis=1) # Calculate non-null counts per row
coint_test_df = coint_test_df.sort_values(by='non_null_count', ascending=False) # Sort in descending order
coint_test_df = coint_test_df.drop('non_null_count', axis=1)
```

```python
# Create the VAR model
model = VAR(endog = diff_data_df
           )  # or VAR(data) if your series were already stationary

# Select optimal lag order
lag_results = model.select_order(trend='n')
# print(lag_results.summary())
```

```python
model.select_order(trend='c').selected_orders
```

```python
lag_number = 4#lag_results.selected_orders['aic']
```

```python
# FULL data VAR maxlags=4, trend='n'

# Fit the model with the selected lag order
var_model = model.fit(maxlags=lag_number
                      , ic=None
                      , trend='c'
                     )

# Summary of the model
print(var_model.summary())
```

```python
forecast_steps_val = 16

point, lower, upper = var_model.forecast_interval(y = diff_data_df.values
                                                  , steps = forecast_steps_val
                                                  , alpha=0.05
                                                 )

point_df = pd.DataFrame(point, columns = diff_data_df.columns
                       )

lower_df = pd.DataFrame(lower, columns = diff_data_df.columns
                       )

upper_df = pd.DataFrame(upper, columns = diff_data_df.columns
                       )
```

```python
point_df.index = pd.date_range(start=max(diff_data_df.index), periods=forecast_steps_val + 1, freq='Q')[1:]
lower_df.index = pd.date_range(start=max(diff_data_df.index), periods=forecast_steps_val + 1, freq='Q')[1:]
upper_df.index = pd.date_range(start=max(diff_data_df.index), periods=forecast_steps_val + 1, freq='Q')[1:]
```

## Granger

```python
lagvar = 8
```

```python
ssr_based_chi2_df = pd.DataFrame()

for macro_variable in diff_data_df.columns[1:]:
    print("="*10 + macro_variable + "="*10)

    test = grangercausalitytests(diff_data_df[[diff_data_df.columns[0]
                                               , macro_variable]]
                                 , lagvar)

    temp_ssr_based_chi2_df = pd.DataFrame([test[1][0]['ssr_chi2test'][1]] + [test[2][0]['ssr_chi2test'][1]] + [test[3][0]['ssr_chi2test'][1]] + [test[4][0]['ssr_chi2test'][1]] + [test[5][0]['ssr_chi2test'][1]]
                                      + [test[6][0]['ssr_chi2test'][1]] + [test[7][0]['ssr_chi2test'][1]] + [test[8][0]['ssr_chi2test'][1]]
                                      , columns=[macro_variable]).T

    ssr_based_chi2_df = pd.concat([ssr_based_chi2_df, temp_ssr_based_chi2_df]) 
    print("\n"*5)
```

```python
ssr_based_chi2_df = ssr_based_chi2_df.T
ssr_based_chi2_df.index = ssr_based_chi2_df.index+1
ssr_based_chi2_df = ssr_based_chi2_df.reset_index(names='Кол-во лагов', drop=False)
```
